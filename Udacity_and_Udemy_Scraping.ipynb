```python
from selenium import webdriver
import time
from pyudemy import Udemy
```

```python
driver = webdriver.Chrome(ChromeDriverManager().install())
```

```python
driver.get('https://www.udacity.com/courses/all')
```

```python
urls = driver.find_elements_by_css_selector('.card_container__25DrK')
names = driver.find_elements_by_css_selector('.card_overview__D0w4B .card_title__35G97')
levels = driver.find_elements_by_css_selector('.card_stats__AV1N0 .card_level__2HNxe')
durations = driver.find_elements_by_css_selector('.card_stats__AV1N0 .card_duration__1hWII')
skills = driver.find_elements_by_css_selector('.card_details__3VdjA .card_detailContent__2eJIl:first-of-type')
```

```python
links = []
course_names = []
course_levels = []
course_durations = []
course_skills = []
course_urls = []
course_prerequest = []
course_costs = []
course_instructors=[]
for i in range(len(urls)):
    links.append(urls[i].get_attribute('href'))
    course_names.append(names[i].text)
    course_levels.append(levels[i].text)
    course_durations.append(durations[i].text)
for i in range(len(course_skills)):
    course_skills.append(course_skills[i].text)
```

```python
course_prerequest =course_prerequest[0:265]
len(course_prerequest)
```

```python
for i in range(len(urls)):
    ins = []
    time.sleep(2) 
    url = links[i]
    request = driver.get(url)
    try:
        prerequest = driver.find_element_by_css_selector('.degree-info-columns_columnList__D-UoB li:last-of-type h5').text
    except:
        prerequest = 'None'
    try:
        cost = driver.find_element_by_css_selector('._price-card_pricingTemplate__Am1WB ins').text
    except:
        cost = 'None'
    try:
        instructors = driver.find_elements_by_css_selector('.degree-instructors_card__15XLG h5')
    except:
        instructors = driver.find_elements_by_css_selector('.course-instructors_instructorsListItem__3AVxu h3')
    if(len(instructors) >= 1):
        for i in instructors:
            ins.append(i.text)
    else:
        ins.append('None')
    course_prerequest.append(prerequest)
    course_instructors.append(ins)
    course_costs.append(cost)
```

```python
driver.quit()
```

```python
import pandas as pd
courses_df = pd.DataFrame({'Name': course_names,
                          'URL': links,
                           'Level':course_levels,
                           'Durations':course_durations,
                          'Prerequest': course_prerequest,
                           'Instructors':course_instructors,
                          'Cost':course_costs})
# print(courses_df.info())
courses_df
```
